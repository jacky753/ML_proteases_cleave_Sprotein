```python
merops_code_list = []
with open('./testdata/Homo_sapiens_peptidases', 'r') as f:
    row_list = f.read().split("\n")
    print(row_list[0])
    print(row_list[1])
    for i in range(1, len(row_list)):
        row = row_list[i].split("\t")
        print(row[2])
        #if row[7] == "":
        #    print("Locus is empty.")
        #    print("-----------------------------------------------")
        #    continue
        merops_code_list.append(row[2])
        print("----------------------------------------------")
print(merops_code_list)
print(len(merops_code_list))
```

```python
subs_count_list = []
for i in range(len(merops_code_mece)):
    merops_code = [merops_code_mece[i]]
    print("merops_code:" + merops_code_mece[i])
    cur.execute("SELECT uniprot_acc, p1 FROM cleavage where code=(%s);", merops_code)
    
    subs = cur.fetchall()
    for sub in subs:
        print(sub)
    subs_count_list.append(len(subs))
    print(subs_count_list)
    
```

```python
print(subs_count_list)
print(len(subs_count_list))
print(sum(subs_count_list))
```

```python
count = 0
exist_list = []
for i in range(len(subs_count_list)):
    if subs_count_list[i] != 0:
        count = count + 1
        exist_list.append(i)
print(count)
print(exist_list)
```

```python
protease_resistered_subs = []
for i in exist_list:
    protease_resistered_subs.append(merops_code_mece[i])
print(protease_resistered_subs)
```
